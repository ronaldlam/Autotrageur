```python
!pwd
import pandas as pd
import plotly
import plotly.graph_objs as go
```

```python
krwusd_df = pd.read_csv('data/sample_forex/krwusdforexdaily_sample.csv')
ethbithumb_df = pd.read_csv('data/sample_history/ethkrwbithumbdays730.csv')
ethgemini_df = pd.read_csv('data/sample_history/ethusdgeminidays730.csv')
```

```python
ethbithumb_df = pd.merge(ethbithumb_df, krwusd_df, on='time', how='inner')
```

```python
# df.iloc[:,[1,2,3,4,7]] *= df['forex_rate']
# df.loc[:,['close']].mul(df['forex_rate'])
ethbithumb_df[["close", "high", "low", "open", "vwap"]] = ethbithumb_df[["close", "high", "low", "open", "vwap"]].multiply(ethbithumb_df["forex_rate"], axis="index")
# from IPython.core.display import HTML
ethbithumb_df
# df[['close','high','low','open','vwap']] = df[['close','high','low','open','vwap']].multiply(df['forex_rate'], axis='index')
# print(df)
```

```python
raw_spread = ethbithumb_df['vwap'] / ethgemini_df['vwap']
raw_spread = ((raw_spread - 1)*100)

# converts unix timestamp to datetime
ethbithumb_df.time = pd.to_datetime(ethbithumb_df.time, unit='s')
ethgemini_df.time = pd.to_datetime(ethgemini_df.time, unit='s')

data = [
    go.Scatter(x=ethbithumb_df.time, y=ethbithumb_df.vwap, mode='lines+markers', text=raw_spread, name="bithumb"),
    go.Scatter(x=ethgemini_df.time, y=ethgemini_df.vwap, mode='lines+markers', name="gemini")
]
plotly.offline.plot(data, filename="bithumb_gemini_spread.html")
```
